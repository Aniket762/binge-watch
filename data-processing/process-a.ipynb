```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('movie_metadata.csv')
data.head()
```

```python
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
data.head()
```

```python
#data cleaning
data['actor_1_name'] = data['actor_1_name'].replace(np.nan,'unkown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan,'unkown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan,'unkown')
data['director_name'] = data['director_name'].replace(np.nan,'unkown')
data['genres'] = data['genres'].replace('|',' ')
data['movie_title'] = data['movie_title'].str.lower()
data.head()
```

```python
data['movie_title'][1]
```

```python
#we remove the \xa0 character , data cleaning
data['movie_title'] = data['movie_title'].apply(lambda x:x[:-1])
data['movie_title'][1]
```

```python
data.to_csv('data.csv',index=False)
```
